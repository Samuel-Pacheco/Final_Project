```python
import pandas as pd
```

```python
df = pd.read_csv('Resources/Meteorite_Landings.csv')
df.head()
```

```python
# Show null values
df.isnull().sum()
```

```python
# Remove null values
df2= df.dropna()
df2.head()
```

```python
df2.isnull().sum()
```

```python
# Rename columns
df3 = df2.rename(columns = {'name': 'meteorite_name', 'mass (g)': 'mass_g', 'year': 'impact_year'})
df3.head()
```

```python
df4 = df3.drop(['id', 'nametype', 'fall', 'reclat', 'reclong','GeoLocation'], axis=1)
df4
```

```python
df4['mass_g'] = (df4['mass_g'].str.replace(',', '').astype(float))
df4
```

```python
# write formatted to csv
df4.to_csv("Meteorite_info_Data.csv")
```

```python
df5 = df3.drop(['id', 'nametype', 'fall', 'mass_g', 'impact_year','GeoLocation', 'impact_year', 'recclass'], axis=1)
df5
```

```python
# write formatted to csv
df5.to_csv("Meteorite_location_Data.csv")
```
