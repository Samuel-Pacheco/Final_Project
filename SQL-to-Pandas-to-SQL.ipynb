```python
db_string = f"postgresql://postgres:{db_password}@127.0.0.1:5432/FinalProject"
```

```python
engine = create_engine(db_string)
conn = engine.connect()
```

```python
M_info_df = pd.read_sql("select * from meteorite_info", conn)
M_info_df
```

```python
M_location_df = pd.read_sql("select * from meteorite_location", conn)
M_location_df
```

```python
## Join in pandas
Meteorite_final_df = pd.merge(M_info_df,
                                    M_location_df,
                                    how='left',
                                    on='meteorite_name')
Meteorite_final_df = Meteorite_final_df[['meteorite_name','recclass','mass_g', 'impact_year', 'retclat','reclong']]
Meteorite_final_df
```

```python
## Join in SQL
Meteorite_final_df = pd.read_sql('''select r.meteorite_name, recclass, mass_g, impact_year, retclat,reclong
                                       from meteorite_info r
                                       left join meteorite_location p
                                           on r.meteorite_name = p.meteorite_name
                                       ''',
                                      con=conn)
Meteorite_final_df 
```
